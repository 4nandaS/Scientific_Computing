```python
# nomor 2 bagian B

import sympy as sp
import math

x = sp.symbols('x')

sin_taylor = sp.sin(x).series(x, 0, 5).removeO()
cos_taylor = sp.cos(x).series(x, 0, 5).removeO()
sin_cos_taylor = (sin_taylor*cos_taylor).expand()

x_value = math.pi/2
separate = sp.sin(x_value) * sp.cos(x_value)
first = sin_cos_taylor.subs(x, x_value)

error_separate = abs(separate - separate.evalf())
error_first = abs(separate - first.evalf())

print("Kesalahan perhitungan terpisah : ", error_separate)
print("Kesalahan perhitungan dengan urutan perkalian : ", error_first)
```
