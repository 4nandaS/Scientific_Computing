```python
# nomor 2 bagian A

import math

def func_sin(x):
    sin_approx = math.sin(x)
    return sin_approx

def func_cos(x):
    cos_approx = math.cos(x)
    return cos_approx

x = 0
out_cos = func_cos(x)
out_sin = func_sin(x)
out_sin_cos = func_sin(x) * func_cos(x)
print("sin(x)", out_sin)
print("cos(x)", out_cos)
print("sin(x)cos(x)", out_sin_cos)
```
