```python
# nomor 1 bagian D

import numpy as np
import matplotlib.pyplot as plt

#plt.style.use('seaborn-poster')

x = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
y = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])
z = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])

# Interpolate the data
y_interpolated = np.interp(z, x, y)

plt.figure(figsize=(15, 15))
plt.plot(x, y, '-ob')
plt.plot(z, y_interpolated, 'ro')

for i in range(len(z)):
    print("Year:", z[i], "Temperature:", y_interpolated[i])

plt.title("1981 - 1999")
plt.xlabel("TAHUN")
plt.ylabel("SUHU")
plt.show()
```
