```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat data koordinat untuk gambar gunung
x = np.linspace(-np.pi, np.pi, 1000)
y = np.sin(x)*np.exp(-0.1*x)

# Mengatur ukuran dan warna gambar
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_facecolor('#123')

# Membuat plot kurva untuk gambar gunung
ax.plot(x, y, color='#124', linewidth=3)

# Menambahkan judul dan menghilangkan sumbu koordinat
ax.set_title('JUDUL', fontsize=20)
ax.set_xticks([])
ax.set_yticks([])

# Menampilkan gambar
plt.show()
```
