```python
import numpy as np
from scipy.integrate import trapz
a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)
I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 *sum(f[1:n-1]) + f[n-1])
print(I_trapz)
print(I_trap)
```

```python
from scipy.integrate import quad  
I_quad, est_err_quad = \
           quad(np.sin, 0,np.pi)
print(I_quad)
err_quad = 2 - I_quad
print(est_err_quad, err_quad)
```
